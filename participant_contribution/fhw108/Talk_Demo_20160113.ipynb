```python
# Save token outside of version control
fh = open('maintoken.txt','r')
mytoken = fh.next().strip()
fh.close()
```

## Export Records

```python
#!/usr/bin/env python
import pycurl, cStringIO
buf = cStringIO.StringIO()
data = {
    'token': mytoken,
    'content': 'record',
    'format': 'csv',
    'type': 'flat',
    'rawOrLabel': 'raw',
    'rawOrLabelHeaders': 'raw',
    'exportCheckboxLabel': 'false',
    'exportSurveyFields': 'false',
    'exportDataAccessGroups': 'false',
    'returnFormat': 'json'
}
ch = pycurl.Curl()
ch.setopt(ch.URL, 'https://redcap2.nubic.northwestern.edu/redcap/api/')
ch.setopt(ch.HTTPPOST, data.items())
ch.setopt(ch.WRITEFUNCTION, buf.write)
ch.setopt(ch.SSL_VERIFYPEER,0)
ch.perform()
ch.close()
#print buf.getvalue()
myval = pd.read_csv(StringIO.StringIO(buf.getvalue()))
buf.close()
myval
```
